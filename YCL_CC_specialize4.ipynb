```python
import tensorflow as keras
import gymnasium as gym
import numpy as np
import os
from keras.utils import to_categorical

def main():
    # Assuming data is loaded from Google Drive or uploaded manually
    model_path = '/content/drive/My Drive/YOURMODEL.h5'  # Update this path
    test_data_path = '/content/drive/My Drive/test.npz'  # Update this path
    train_data_path = '/content/drive/My Drive/train.npz'  # Update this path

    model = keras.models.load_model('/content/YOURMODEL.h5')
    val_dataset = np.load(test_data_path)
    val_data = val_dataset['data']
    val_label = to_categorical(val_dataset['label'], num_classes=5)

    train_dataset = np.load(train_data_path)
    train_data = train_dataset['data']
    train_labels = to_categorical(train_dataset['label'], num_classes=5)

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    model.fit(
        train_data,
        train_labels,
        batch_size=32,
        epochs=10,
        validation_data=(val_data, val_label)
    )

    predictions = model.predict(val_data)
    predicted_classes = np.argmax(predictions, axis=1)
    true_labels = np.argmax(val_label, axis=1)
    accuracy = np.mean(true_labels == predicted_classes)
    print(f'Accuracy: {accuracy}')

if __name__ == "__main__":
    main()
```
